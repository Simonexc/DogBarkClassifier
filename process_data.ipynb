```python
import pandas as pd
from pathlib import Path
```

```python
sheet = Path("/home/simonexc/Downloads/5117901/ARCA23K.ground_truth/train.csv")
folder = Path("/home/simonexc/Downloads/5117901/ARCA23K.audio/")
```

```python
files = set(path.name.replace(".wav", "") for path in folder.glob("*.wav"))
files
```

```python
df = pd.read_csv(sheet)
```

```python
df
```

```python
bark = []
no_bark = []
for i, row in df.iterrows():
    if str(row["fname"]) in files:
        if row["label"].lower() == "bark":
            bark.append(str(row["fname"]))
        else:
            no_bark.append(str(row["fname"]))
```

```python
len(bark)
```

```python
len(no_bark)
```

```python
import shutil

for f in files:
    if f in bark:
        shutil.copyfile(folder / f"{f}.wav", Path("data/new/bark") / f"{f}.wav")
    else:
        shutil.copyfile(folder / f"{f}.wav", Path("data/new/no_bark") / f"{f}.wav")
```
